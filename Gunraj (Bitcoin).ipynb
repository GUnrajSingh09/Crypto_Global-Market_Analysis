```python
import mplfinance as mpf
```

```python
mapping = dict(data_details[['Asset_ID', 'Asset_Name']].values)
data['Asset Name'] = data['Asset_ID'].map(mapping)
```

```python
data.head()
```

**ID 1 BITCOIN**

```python
bitcoin = data.query('Asset_ID == 1').reset_index(drop= True) 
bitcoin['timestamp'] = pd.to_datetime(bitcoin['timestamp'], unit = 's') 
bitcoin = bitcoin.set_index('timestamp')
```

```python
bitcoin
```

```python
whole = bitcoin.loc['2021-09-20 22:59:00': '2021-09-21 00:00:00']
```

```python
mpf.plot(
    whole,
    type='candle',
    volume = True,
    mav=(3,6,9),
    figratio=(3,1),
    style='yahoo',
    title='Bitcoin'
)
```

```python
mpf.plot(
    whole,
    type='line',
    volume = True,
    
    
    title='Bitcoin'
)
```

```python
mpf.plot(
    whole,
    type='candle',
    volume = True,
    
    style = 'yahoo',
    title='Bitcoin'
)
```

REAL TIME USING YAHOO FINANCE API

```python
!pip -q install yfinance
import yfinance as yf

bitcoin_now = yf.Ticker('BTC-USD')

bitcoin_now_values = bitcoin_now.history(start='2021-09-21')

mpf.plot(
    bitcoin_now_values,
    type='candle',
    volume = False,
    mav=(3,6,9),
    figratio=(3,1),
    style='yahoo',
    title='bitcoin'
)
```
