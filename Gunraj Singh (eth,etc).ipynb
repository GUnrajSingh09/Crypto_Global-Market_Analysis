```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../input/g-research-crypto-forecasting/train.csv")
data_details = pd.read_csv("../input/g-research-crypto-forecasting/asset_details.csv")
```

```python
!pip -q install mplfinance
```

```python
import mplfinance as mpl
mapping = dict(data_details[['Asset_ID', 'Asset_Name']].values)
data['Asset_Name'] = data['Asset_ID'].map(mapping)
```

```python
data.head(14)
```

# **ID 6. Ethereum (eth)**

```python
eth = data.query('Asset_ID == 6').reset_index(drop = True)
eth['timestamp'] = pd.to_datetime(eth['timestamp'], unit = 's')
eth = eth.set_index('timestamp')
```

```python
eth
```

```python
whole = eth.loc['2021-09-20 18:00:00': '2021-09-21 00:00:00']
```

```python
mpl.plot(
    whole,
    type='candle',
    volume = True,
    mav=(50,100,200),
    figratio=(3,1),
    style='yahoo',
    title='Ethereum (eth)'
)
```

```python
!pip -q install yfinance
```

```python
import yfinance as yf

eth_now = yf.Ticker('ETH-USD')

eth_now_values = eth_now.history(start='2021-09-21')

mpl.plot(
    eth_now_values,
    type='candle',
    volume = True,
    mav=(15,30,50),
    figratio=(3,1),
    style='yahoo',
    title='Ethereum (eth)'
)
```
